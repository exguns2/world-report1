# Visualization

Bu gÃ¶rselleÅŸtirme, regresyon modelinde hangi deÄŸiÅŸkenlerin en gÃ¼Ã§lÃ¼ tahmini gÃ¼ce sahip olduÄŸunu ve bu iliÅŸkilerin doÄŸasÄ±nÄ± hÄ±zlÄ±ca anlamamÄ±zÄ± saÄŸlar.

```python
def pltscatter(var1,var2 = 'Ladder score',df = data1):
    df.plot(kind='scatter', x=var1, y=var2,alpha = 0.5,color = 'red')
    plt.xlabel(var1)              # label = name of label
    plt.ylabel(var2)
```

```python
cat1=['Log GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption','Prosperity Index']
for c in cat1:
    pltscatter(c)
```

Bu gÃ¶rselleÅŸtirme ile, her bir faktÃ¶rÃ¼n ne kadar homojen ya da heterojen daÄŸÄ±ldÄ±ÄŸÄ± netleÅŸir.

```python
def plot_histogram(variable, df=data1):

    plt.figure(figsize=(8, 5))
    sns.histplot(
        data=df,
        x=variable,
        bins=15,
        color='skyblue'
    )
    plt.title(f'{variable} DeÄŸerlerinin Frekans DaÄŸÄ±lÄ±mÄ±')
    plt.xlabel(variable)
    plt.ylabel('Frekans')
    plt.show()
```

```python
for c in cat1:
    plot_histogram(c)
```

BÃ¶lgesel karÅŸÄ±laÅŸtÄ±rma, deÄŸiÅŸkenlere gÃ¶re bir Ã¼lkenin refah seviyesi Ã¼zerindeki belirleyici rolÃ¼nÃ¼ aÃ§Ä±kÃ§a gÃ¶zler Ã¶nÃ¼ne sermektedir.

```python
def barplot(var1,var2='Regional indicator',df= data1):
    plt.figure(figsize=(14, 7))
    sns.barplot(
        data=data1, 
        x=var2,
        y=var1,)
    
    plt.xticks(rotation=45, ha='right')
    plt.title(f'{var1} vs. {var2}', fontsize=14)
    plt.xlabel(var2)
    plt.ylabel(var1)
    plt.show()
    
for c in cat1:
    barplot(c)
```

Bu kutu grafiÄŸi, `Ladder score` deÄŸiÅŸkeninin, `Regional indicator` kategorisine gÃ¶re nasÄ±l daÄŸÄ±ldÄ±ÄŸÄ±nÄ± gÃ¶rselleÅŸtirmektedir. Bu gÃ¶rselleÅŸtirme, sadece hangi bÃ¶lgelerin ortalama olarak daha mutlu olduÄŸunu deÄŸil, aynÄ± zamanda o bÃ¶lge iÃ§inde mutluluk seviyelerinin ne kadar tutarlÄ± (dar kutu) veya dengesiz (geniÅŸ kutu) olduÄŸunu da anlamamÄ±zÄ± saÄŸlar. 

```python
def box(var1,var2='Prosperity Level',df=data1):
    df.boxplot(column=var1,by = var2)
for c in cat1:
    box(c)
```

Bu daÄŸÄ±lÄ±m grafiÄŸi, bu iki deÄŸiÅŸken arasÄ±ndaki genel korelasyonu gÃ¶sterir. 

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=data1, 
    x='Perceptions of corruption',
    y='Ladder score',        
    alpha=0.6,
    color='darkblue')

plt.title("Ladder Score vs. Perceptions of corruption", 
          fontsize=14, fontweight='bold')
plt.xlabel('Perceptions of corruption', fontsize=12)
plt.ylabel('Ladder Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Ladder score iÃ§in IQR hesaplama

```python
Q1 = data1['Ladder score'].quantile(0.25)
Q3 = data1['Ladder score'].quantile(0.75)
IQR = Q3 - Q1

iqr_l = Q1 - 1.5 * IQR
iqr_h = Q3 + 1.5 * IQR
```

Bu analizde, `Ladder score` daÄŸÄ±lÄ±mÄ±nÄ±n **en uÃ§ noktalarÄ±** ile (`Dystopia + residual`) arasÄ±ndaki iliÅŸki incelenecektir.

**AmaÃ§:** Zaten aÅŸÄ±rÄ± mutlu/mutsuz olan Ã¼lkeler arasÄ±nda, modelin beklentisini en Ã§ok aÅŸan (en yÃ¼ksek pozitif residual) ve en Ã§ok hayal kÄ±rÄ±klÄ±ÄŸÄ±na uÄŸratan (en yÃ¼ksek negatif residual) vakalarÄ± tespit etmektir.

```python
upper_outliers = data1[data1['Ladder score'] > iqr_h].copy()

upper_outliers_sorted = upper_outliers.sort_values(
    by='Dystopia + residual', 
    ascending=False)

print("\n ðŸŒŸ Beklenenden En Mutlu")
print(upper_outliers_sorted[['Country name', 'Ladder score', 'Dystopia + residual']].head(5).to_markdown(index=False))

lower_outliers = data1[data1['Ladder score'] < iqr_l].copy()

lower_outliers_sorted = lower_outliers.sort_values(
    by='Dystopia + residual', 
    ascending=True)

print("\nðŸ˜ž Beklenenden En Mutsuz")
print(lower_outliers_sorted[['Country name', 'Ladder score', 'Dystopia + residual']].head(5).to_markdown(index=False))
```
